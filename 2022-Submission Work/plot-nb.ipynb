```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
# set width of bars
barWidth = 0.25
 
# set heights of bars
builtin = [92,80,75]
sc= [84,75,70]
 
# Set position of bar on X axis
r1 = np.arange(len(builtin))
r2 = [x + barWidth for x in r1]

 
# Make the plot
plt.bar(r1, builtin,  color='orange',width=barWidth, edgecolor='white', label='Built-in Function')
plt.bar(r2, sc, color='dodgerblue', width=barWidth, edgecolor='white', label='Blockchain Smart Contract')

 
# Add xticks on the middle of the group bars
plt.xlabel('Data Type', fontweight='bold',fontsize=14)
plt.ylabel('Prediction Accuracy in Percent(%)', fontweight='bold',fontsize=14)
plt.xticks([r +0.1  for r in range(len(builtin))], ['Iris Flower', 'Pima Diabetes', 'Heart Disease'],fontsize=13)
plt.yticks(fontsize=13)
# Create legend & Show graphic

plt.legend()
plt.savefig("NB-Accuracy.pdf")
plt.show()
```
